```python
%matplotlib inline
import matplotlib.pyplot as plt
import analyzeWiggles as aw
import pandas as pd
import numpy as np
```

```python
all_means_weekday = !ls ../final/data/weekday/total_flow_weekday_mean_vector.pivot_*
all_means_weekend = !ls ../final/data/weekend/total_flow_weekend_mean_vector.pivot_* 

fig = plt.figure(figsize = (20,30))

####
ax = fig.add_subplot(3, 2, 1)
plt.title("Smoothed, Cubic 12")
legendItems = []
for a in all_means_weekday:
    vectors = aw.smooth(a, 'cubic', 12)
    plt.plot(vectors['timelabels'], vectors['smoothedVector'])
    legendItems.append(a[a.find('grouping')-5:a.find('grouping')-1])
plt.legend([a for a in legendItems])

####
ax = fig.add_subplot(3, 2, 3)
plt.title("Smoothed, Cubic 8")
legendItems = []
for a in all_means_weekday:
    vectors = aw.smooth(a, 'cubic', 8)
    plt.plot(vectors['timelabels'], vectors['smoothedVector'])
    legendItems.append(a[59:63])
plt.legend([a for a in legendItems])

####
ax = fig.add_subplot(3, 2, 5)
plt.title("Wiggle Magnitude")
legendItems = []
for a in all_means_weekday:
    vectors = aw.smooth(a, 'cubic', 8)
    plt.plot(vectors['timelabels'],vectors['diffVector'])
    legendItems.append(a[59:63])

for i in range(0,288,12):
    plt.axvline(x=i, linewidth=.1, color='gray')

ax = fig.add_subplot(3, 2, 2)
plt.title("Smoothed, Cubic 12")
legendItems = []
####
for a in all_means_weekend:
    vectors = aw.smooth(a, 'cubic', 12)
    plt.plot(vectors['timelabels'], vectors['smoothedVector'])
    legendItems.append(a[59:63])
plt.legend([a for a in legendItems])

####
ax = fig.add_subplot(3, 2, 4)
plt.title("Smoothed, Cubic 8")
legendItems = []
for a in all_means_weekend:
    vectors = aw.smooth(a, 'cubic', 8)
    plt.plot(vectors['timelabels'], vectors['smoothedVector'])
    legendItems.append(a[59:63])
plt.legend([a for a in legendItems])

####
ax = fig.add_subplot(3, 2, 6)
plt.title("Wiggle Magnitude")
legendItems = []
for a in all_means_weekend:
    vectors = aw.smooth(a, 'cubic', 8)
    plt.plot(vectors['timelabels'],vectors['diffVector'])
    legendItems.append(a[59:63])

for i in range(0,288,12):
    plt.axvline(x=i, linewidth=.1, color='gray')

plt.show()
```
